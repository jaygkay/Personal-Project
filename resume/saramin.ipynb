#### main.py 완성
- 사람인 채용사이트의 빅히트 직무 다 긁어옴
- 데이터 정제 및 생성

```python
last_page = str(soup.find_all('a', {'class' :'last'}))
last_page = int(last_page.split('page=')[1][:2])
last_page
```

```python
job_lst = []
url_lst = []
jid_lst = []
com_lst = []

for page_num in range(1, last_page + 1):

    client = requests.session()
    url = "https://bighit.saramin.co.kr/apply_site/recruit/list?page={}".format(page_num)
    response = client.get(url)
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')
    
    # company, job, job_id, url
    for i in soup.select(".recruit_view"):
        com_lst.append(str(i.text).split(']')[0][2:])
        job_lst.append(str(i.text).split(']')[1])
        url_lst.append(i['href'])
        jid_lst.append(int(str(i['href']).split('/')[-1]))
    
                
df = pd.DataFrame({'company':com_lst, 'job_id':jid_lst, 'job':job_lst, 'url':url_lst,})
df
```
